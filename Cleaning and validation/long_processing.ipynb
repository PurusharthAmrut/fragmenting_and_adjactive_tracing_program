```python
import pandas as pd
import re
from langdetect import detect
from langdetect import DetectorFactory

DetectorFactory.seed = 0

data = pd.read_csv('cleaned_dataset.csv')
data.head(8)
```

```python
data.dropna(inplace=True)
```

```python
sum(data['Review Votes'].isnull())
```

```python
data.reset_index(inplace=True)
```

```python
reference = data['Review Votes']>0
index = -1
for row in reference:
    index=index+1
    if not row:
        data.drop(index,inplace=True)
print('Program finished!')
```

```python
data.reset_index(inplace=True)
```

```python
index = -1
try:
    for review in data['Reviews']:
        index = index + 1
        if re.search(r'[A-Za-z]',review) is None:
            data.drop(index,inplace=True)
            continue
        elif re.search(r'^http.*$',review) is not None:
            data.drop(index,inplace=True)
            continue
        elif len(review.split())<=2:
            continue
        if detect(review)!='en':
            data.drop(index,inplace=True)
        if index%1000==0:
            print('Reached',index)
finally:
    print('last index:',index)
```
